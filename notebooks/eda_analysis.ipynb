# Privacy Policy Grader — EDA & Analysis Notebook

This notebook provides exploratory data analysis of graded privacy policies.
Run `python download_data.py` and `python pipeline.py` before using this notebook.

```python
import os
import json
import glob
from pathlib import Path

# Load all result JSONs
results_dir = Path('../results')
results = []

for f in sorted(results_dir.glob('*.json')):
    with open(f) as fh:
        data = json.load(fh)
        data['_file'] = f.name
        results.append(data)

print(f'Loaded {len(results)} result files')
for r in results:
    print(f"  {r.get('name', r['_file'])}: Grade={r.get('grade')}, Score={r.get('final_risk_score')}")
```

```python
# Risk score distribution
import collections

grade_counts = collections.Counter(r.get('grade', 'N/A') for r in results)
print('Grade distribution:')
for grade in ['A', 'B', 'C', 'D', 'F']:
    count = grade_counts.get(grade, 0)
    bar = '█' * count
    print(f'  {grade}: {bar} ({count})')
```

```python
# Red flags frequency
from collections import Counter

all_flags = []
for r in results:
    all_flags.extend(r.get('red_flags', []))

print('Top red flags across all policies:')
for flag, count in Counter(all_flags).most_common(10):
    print(f'  [{count}] {flag}')
```

```python
# Data collected types
all_data_types = []
for r in results:
    all_data_types.extend(r.get('data_collected', []))

print('Data types collected (most common):')
for item, count in Counter(all_data_types).most_common(10):
    print(f'  [{count}] {item}')
```

```python
# Full details for one policy
import pprint

if results:
    print(f"=== Full Result: {results[0].get('name', 'First')} ===")
    pprint.pprint(results[0])
```

```python
# Check processed data chunks
data_dir = Path('../data')
data_files = sorted(data_dir.glob('*.json'))
print(f'Data files: {len(data_files)}')

for f in data_files:
    with open(f) as fh:
        d = json.load(fh)
    print(f"  {f.name}: {d.get('num_chunks')} chunks, raw_length={d.get('raw_length', 'N/A')}")
```
